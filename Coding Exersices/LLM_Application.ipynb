# Knock-Knock Chatbot
>The goal of this notebook is to present a design for an application that uses an LLM. It will showcase human and AI interaction; and demonstrate an understanding of LLMs through the application engagement elements with users to promote learning, culture, and inclusion.

**The goal of this application is to provide users  with various knock-knock jokes at a global scale. It is also tool that can explain the joke outputted.  This will bring joy to the world! All while providing users with a slightly academic yet super fun tool.**

## Example of a Human to  Ai interaction on Knock-Knock

>**User**: English ***Provides language***
>
>**Ai**: Knock-Knock 
>
>**User**: Who’s there? ***Accepts variations of “Who’s there?”***
>
>**Ai**: Water. ***Initiates Joke***
>
>**User**: Water who? ***Accepts variations of  “[Joke] who?”***
>
>**Ai**: Water you doing today? ***Joke Punchline***
>
>**User**: Can you explain the joke to me? ***User asks for an explanation.***
>
>**Ai**: This is a play on words for “What are you doing here?”  instead “Water doing here?”. It’s a pun. ***In simple language the Ai generates a response***

Joke taken from [Goodhousekeeping.com](https://www.goodhousekeeping.com/uk/lifestyle/g62879688/knock-knock-jokes/) 


##### This example of how Knock-Knock would function showcases the interaction between humans and Ai and demonstrates LLM functions. 

+ **Generative function:** Joke generation uses patterns and structures of human language to create new jokes. 

+ **Adaptive**: Ability to response to different variations of user input. Also having the ability to respond in different languages. 


##### This application also showcases the ability for augment learning, and engagement with culture. 

+ **Augment learning:** This model focuses on wordplay and other  literary techniques where users interacting can engage with reading and comprehension skills. It’s focus on language also proves useful to those studying new languages, 

+ **Enhance engagement with culture:** This model is combatable with multiple languages. This is a direct engagement with culture as humour plays a large role in any cultural dynamic. 

+ **Inclusion**: This model is combatable with multiple languages. This reflects diversity in culture. This also is important because it provides a safe space where offensive stereotypes should not in play (*see implications for more information*)



## Reflection
While there is a lot of positives of this application, there is also potential for a lot of social and cultural implications. 
+ **Sensitivity**: 
    + ***Culturally***: Jokes about one specific culture can be confusing to another. It’s important that the AI model trains on understanding language to culture model to the best of its ability. 
    + ***Socially***: Jokes can sometimes single people out. If a user is sensitive to certain jokes, they may get offended, so it’s important for AI to train on datasets that are not providing these kinds of examples. Or training on data that explain why these jokes are offensive. 
 + **Offensive Content**:  Humour can sometimes harm people by being politically incorrect or discriminatory. It’s important to include content filters for sensitive keywords and topics. 


>To mitigate all these issues there are a plethora of ways stated in the text but all negative implications that are not caught by mitigation strategies should be documented using a reporting system.
