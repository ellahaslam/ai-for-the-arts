> **Real-World Example: Analyzing Student Scores with NumPy**
>Imagine you have a list of student scores and you want to:
>+ Calculate the average score
>+ Find the highest and lowest scores
>+ Apply a curve (increase all scores by 5%)

```python
import numpy as np

# Student scores
scores = np.array([85, 90, 78, 92, 88])

# 1. Calculate average score
average = np.mean(scores)

# 2. Find highest and lowest scores
highest = np.max(scores)
lowest = np.min(scores)

# 3. Apply a curve (increase all scores by 5%)
curvedscores = scores * 1.05

# Display results
print(f"Original scores: {scores}")
print(f"Average score: {average}")
print(f"Highest score: {highest}")
print(f"Lowest score: {lowest}")
print(f"Curved scores: {curvedscores}")
```

The else statement acts as a default case in an if-elif-else structure. Here’s the reasoning:
+ When you write if and elif conditions, you are checking for specific cases (e.g., "happy" or "sad").

+ Neither if mood == "happy" nor elif mood == "sad" will be true.
+ Without else, the program would do nothing in that situation, which might confuse the user.
+ By adding else, we ensure the program always responds, even if the input doesn’t match any of the earlier conditions.

```python

import sys

# Multilingual messages for inclusivity
MESSAGES = {
    "en": "Hello, World!",
    "es": "¡Hola, Mundo!",
    "fr": "Bonjour, le monde!",
    "de": "Hallo, Welt!"
}

def sanitizeinput(text: str) -> str:
    """
    Sanitize input to prevent injection in HTML contexts.
    Escape &, <, and > in the raw text.
    """
    return (
        text.replace("&", "&amp;")
            .replace("<", "&lt;")
            .replace(">", "&gt;")
    )

def normalize_language(text: str) -> str:
    """
    Normalize language input to a safe token (letters only),
    then use it as a key.
    """
    # Keep only a–z characters; lower-case it
    normalized = "".join(ch for ch in text.lower() if "a" <= ch <= "z")
    return normalized or "en"

def greet(language="en"):
    """
    Print greeting in the chosen language.
    Falls back to English if language is unsupported.
    """
    key = normalize_language(language)
    message = MESSAGES.get(key, MESSAGES["en"])
    print(message)

if __name__ == "__main__":
    # Default language is English
    lang = "en"

    # Check if user provided a language via command-line argument
    if len(sys.argv) > 1:
        # Sanitize (if later echoed into HTML) and normalize for key lookup
        raw_arg = sys.argv[1].strip()
        lang = normalize_language(raw_arg)
    else:
        # Ask user for language choice interactively
        userinput = input("Choose language (en/es/fr/de): ").strip()
        lang = normalize_language(userinput)

    greet(lang)
```

```python

def bake_cupcakes(ingredients, oven_temperature_c):
    """
    Bake cupcakes following the described steps.

    Parameters
    ----------
    ingredients : list[str]
        A list of ingredient names (e.g., ["flour", "sugar", "eggs", "butter"])
    oven_temperature_c : int | float
        The temperature you set the oven to (in °C)

    Returns
    -------
    str
        "Cupcakes are made" when the process completes.
    """

    # 1) Preheat oven to 180C
    print(f"Preheating oven to {oven_temperature_c}°C...")
    if oven_temperature_c != 180:
        print("Tip: Recipe expects 180°C; adjust if needed.")

    # 2) Create a list 'bowl' and add ingredients
    bowl = []
    print("Creating bowl and adding ingredients:")
    for ingredient in ingredients:
        bowl.append(ingredient)
        print(f"  - Added {ingredient}")

    # 3) If all ingredients are in bowl, mix until smooth batter
    if len(bowl) == len(ingredients):
        print("All ingredients are in the bowl. Mixing until smooth batter is created...")
        batter = "smooth batter"
    else:
        raise ValueError("Not all ingredients were added to the bowl.")

    # 4) Fill cupcake tray with batter
    print("Filling cupcake tray with batter from bowl...")
    cupcake_tray = ["batter"] * 12  # simple representation of a 12-cup tray

    # 5) Put into oven and bake for 35 minutes
    print("Putting tray into oven...")
    bake_time_minutes = 35
    print(f"Baking for {bake_time_minutes} minutes at {oven_temperature_c}°C...")

    # --- Simple doneness check (placeholder logic) ---
    # In a real scenario you might check internal temperature or use a skewer test.
    # Here we simulate "not fully cooked" if temperature < 180.
    cupcakes_fully_cooked = oven_temperature_c >= 180

    # 6) If cupcakes are not fully cooked, bake 5 more minutes; else remove tray
    if not cupcakes_fully_cooked:
        print("Cupcakes are not fully cooked. Baking for 5 more minutes...")
        bake_time_minutes += 5
        cupcakes_fully_cooked = True  # assume the extra time finishes them
    else:
        print("Cupcakes look done—removing tray from oven.")

    print(f"Total bake time: {bake_time_minutes} minutes.")
    print("Removing tray from oven...")
    print("Cupcakes are made")

    return "Cupcakes are made"


# --- Example usage ---
if __name__ == "__main__":
    ingredients_list = ["flour", "sugar", "eggs", "butter", "milk", "vanilla"]
    result = bake_cupcakes(ingredients_list, 180)
```
