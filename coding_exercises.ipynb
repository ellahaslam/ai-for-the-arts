## Functions
### Initial Prompt
Could you suggest a modification of the Python code for “Hello, world!” to demonstrate the concept of functions to a complete beginner in programming? Please explain the code step-by-step highlighting any key aspects. 

### Code

```python
def greet(): 
    print("Hello, world!") 
greet() 
```

### Suggested Prompt
How to return a value from a function instead of just printing?  

### Code 

```python
def greet(): 
    return "Hello, world!" 
message = greet() 
print(message) 
```

### My Prompt
Why would you use greet instead of print? Is it not the same thing? 
>  My thought process on this question was I did not understand there are purposes for functions that go beyond the print function if both are displaying text. 

### Copilot Response and Code
Why define greet() then? 

1. Reusability 

If you need to print "Hello, world!" multiple times, you can just call greet() instead of rewriting print("Hello, world!") everywhere. 

Example: 

```python
greet() 
greet() 
```

2. Customization 

You can later add parameters to make the function more flexible:  

```python
def greet(name): 
    print(f"Hello, {name}!") 
# Now you can greet anyone:  
greet("Jane Doe") 
greet("John Doe") 
```

3. Organization 
+ Functions help structure your code into logical blocks. 
+ In larger programs, you might have many tasks (e.g., greeting, calculating, saving data). Functions keep these tasks separate and easy to manage. 

4. Scalability 
+ Today, greet() just prints text. Tomorrow, you might want it to: 
    + Log the greeting to a file 
    + Add a timestamp 
    + Return the greeting instead of printing it 
+ With a function, you can change its behavior without touching every place where it’s used. 

### Key Learning Outcomes
+ Offers an explanation as to why returning values from a function are important and why return values are used. 
> **From Copilot** : Great question! Returning a value from a function is an important concept because it allows you to use the result elsewhere in your program, instead of just displaying it.
