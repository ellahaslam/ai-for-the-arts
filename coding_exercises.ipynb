> **Real-World Example: Analyzing Student Scores with NumPy**
>Imagine you have a list of student scores and you want to:
>+ Calculate the average score
>+ Find the highest and lowest scores
>+ Apply a curve (increase all scores by 5%)

```python
import numpy as np

# Student scores
scores = np.array([85, 90, 78, 92, 88])

# 1. Calculate average score
average = np.mean(scores)

# 2. Find highest and lowest scores
highest = np.max(scores)
lowest = np.min(scores)

# 3. Apply a curve (increase all scores by 5%)
curvedscores = scores * 1.05

# Display results
print(f"Original scores: {scores}")
print(f"Average score: {average}")
print(f"Highest score: {highest}")
print(f"Lowest score: {lowest}")
print(f"Curved scores: {curvedscores}")
```

### Response and Code
The else statement acts as a default case in an if-elif-else structure. Here’s the reasoning:
+ When you write if and elif conditions, you are checking for specific cases (e.g., "happy" or "sad").
+ But what if the user types something unexpected like "excited" or "angry"?
+ Neither if mood == "happy" nor elif mood == "sad" will be true.
+ Without else, the program would do nothing in that situation, which might confuse the user.
+ By adding else, we ensure the program always responds, even if the input doesn’t match any of the earlier conditions.
